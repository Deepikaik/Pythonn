```python
#normal for loop
num=[10,20,30,40,50]
newnum=[]
for x in num:
    newnum.append(x*x)
print(newnum)
```

```python
#map fnc with lambda
num=[10,20,30,40,50]
newnum=list(map(lambda num:num*num,num))
print(newnum)
```

```python
#map fnc
num=[10,20,30,40,50]
def newnumfnc(num):
    return num*num
newnum=list(map(newnumfnc,num))
print(newnum)
```
